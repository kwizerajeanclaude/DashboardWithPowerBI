# Checking the shape of each DataFrame.
In this step I will be checking the number of rows and the columns of each dataframe and make the total

```python
sales_df = [sales_1,sales_2,sales_3,sales_4,sales_5,sales_6,sales_7,sales_8,sales_9,
                     sales_10,sales_11,sales_12]
```

```python
#counting the rows of the of each dataframe in total before combining them.

for idx, df in enumerate(sales_df, start=1):
    print(f"Shape of DataFrame {idx}: {df.shape}")

#Checking total of all rows

total_rows = 0

for df in sales_df:
    rows, cols = df.shape
    total_rows += rows

print(f"Total number of rows across all DataFrames: {total_rows}")
```

# Renaming columns 
In this step, I am going to rename the columns so that will be easy to merge them

```python

new_cols = {'Order_ID':'ORDERID','Product':'PRODUCT','Quantity_Ordered':'QUANTITY','Price_Each':'PRICE',
            'Order_Date':'DATE','Purchase_Address':'ADDRESS','Order ID':'ORDERID','Quantity Ordered':'QUANTITY',
            'Price Each':'PRICE','Order Date':'DATE','Purchase Address':'ADDRESS'}
```

```python
for sales in sales_df:
    sales.rename(columns=new_cols, inplace=True)
    
```

```python
sales = pd.concat(sales_df, ignore_index=True)
```
