# Feature engineering

```python
sales['QUANTITY'] = pd.to_numeric(sales['QUANTITY'], errors = 'coerce')
sales['PRICE'] = pd.to_numeric(sales['PRICE'], errors = 'coerce')
```

```python
# creation of new column based on question to be answer
sales['Total_Sales'] = sales['QUANTITY']*sales['PRICE']
```

```python
sales.head(10)
```

```python
# separating address into city, state and postal code
sales[['City', 'State', 'Postal Code']] = sales['ADDRESS'].str.split(', ', expand=True)
sales[['City', 'State', 'Postal Code']] = sales[['City', 'State', 'Postal Code']].apply(lambda x: x.str.strip())
sales.drop(columns=['ADDRESS'], inplace=True)
```

```python
sales.head(5)
```

```python
sales.head(10)
```

```python
sales = sales[sales['DATE'].str.contains(r'^\d{2}/\d{2}/\d{2}.*', na=False)]
```

```python
# Time Related
sales['DATE'] = pd.to_datetime(sales['DATE'])

sales['Year'] = sales['DATE'].dt.year
sales['Month'] = sales['DATE'].dt.month
sales['Day'] = sales['DATE'].dt.day
sales['Hour'] = sales['DATE'].dt.hour
```

# Can we identify any seasonality in the sales?

```python
sales_seasonality = sales.groupby('Month')['Total_Sales'].sum()
```

```python
# Visualize sales over months to check for seasonality
plt.plot(sales_seasonality)
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Sales Over Months')
plt.show()
```

# What are our best and worst-selling products?

```python
# Best and worst-selling products
best_selling_products = sales.groupby('PRODUCT')['Total_Sales'].sum().nlargest(5)
worst_selling_products = sales.groupby('PRODUCT')['Total_Sales'].sum().nsmallest(5)
```

```python
best_selling_products
```

```python
worst_selling_products
```

# How do sales compare to previous months or weeks?

```python
monthly_sales = sales.groupby(['Year', 'Month'])['Total_Sales'].sum().reset_index()
monthly_sales['Sales_Change'] = monthly_sales['Total_Sales'].diff()
```

```python
monthly_sales['Sales_Change']
```

# Which cities are our products delivered to most?

```python
# Cities where products are delivered most
top_delivered_cities = sales['State'].value_counts().nlargest(5)
```

```python
#Create a bar plot
plt.figure(figsize=(8, 6))
top_delivered_cities.plot(kind='bar', color='skyblue')

# Adding labels and title
plt.xlabel('State')
plt.ylabel('Number of Deliveries')
plt.title('Top Delivered States')
```

# How do product categories compare in revenue generated and quantities ordered?

```python
product_categories_comparison = sales.groupby('PRODUCT').agg({
    'Total_Sales': 'sum',
    'QUANTITY': 'sum'
})

# Visualize product categories comparison
product_categories_comparison.plot(kind='bar', stacked=True)
plt.xlabel('Product Category')
plt.ylabel('Amount')
plt.title('Product Categories Comparison')
plt.legend(["Revenue", "Quantities Ordered"])
plt.show()
```
